```python
import requests    #importing the requests library, which is used to send HTTP requests and handle the API response.
import json        #imports the json library, which provides functions to work with JSON data.
import pandas as pd #importing the pandas library and assigns it the alias pd. 

# This sends an HTTP GET request to the specified URL and assigns the response to the response_API variable.
#  It retrieves the information about the TV show "Westworld" .
response_API = requests.get('http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes')
# This extracts the JSON data from the API response and parses it into a Python dictionary, and assigned into 'data'.
data = response_API.json()

#This defines a list of attributes that we want to extract from the API response.
attributes = ['id', 'url', 'name', 'season', 'number', 'type', 'airdate', 'airtime', 'airstamp', 'runtime', 'rating', 'summary']

#  extracting the URLs of the medium-sized and original images of the TV show from the image dictionary within the data variable.
image_medium = data['image']['medium']
image_original = data['image']['original']

# retrieving the list of episodes from the _embedded dictionary within the data variable.
episodes = data['_embedded']['episodes']

# This will start a loop to iterate over each episode in the episodes list.
episode_data = []
for episode in episodes:
    episode_info = {field: episode[field] for field in attributes}   #This creates a dictionary called episode_info that contains the attributes for the current episode. It uses a dictionary comprehension to iterate over the attributes list and retrieve the corresponding values from the episode dictionary.
    episode_info['image_medium'] = image_medium
    episode_info['image_original'] = image_original
    episode_data.append(episode_info)

# Create a DataFrame from the episode data
df = pd.DataFrame(episode_data)

# Save DataFrame to a CSV file
df.to_csv('tv_show.csv', index=False)

```
